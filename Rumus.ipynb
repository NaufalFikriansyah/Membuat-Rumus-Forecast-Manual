```python
#Import 
import pandas as pd

import_file = '/content/drive/MyDrive/Project/rumus/DATAOLAH.xlsx' #Sesuaikan sama path drivenya
df = pd.read_excel(import_file, sheet_name='PERAMALAN(2,0,0)') #Ngambil file dari excel yang nama sheet nya PERAMALAN(2,0,0)

df
```

```python
df = df[['yt','ACF']]

df
```

```python
yt = [] #Nampung data yt / data aktual
for i in df['yt']:
  yt.append(i) #masukin nilai aktual kedalam variable yt

y_t = [] #menampung nilay y^t
y_t.append(26)
y_t.append(22)

len_yt = len(yt) 
len_y_t =  len(y_t)

forecast =[] #menampung hasil forecast

f1 = df['ACF'][0]
f2 = df['ACF'][1]
```

RUMUS UNTUK SEBELUM FORECAST

```python
for i in range (0, 95): #index 0 sampai 95
  rumusnya = (f1 * yt[i]) + (f2 * yt[i+1]) #karena index di yt[96] itu index terakhir yakni nilai 30
  y_t.append(rumusnya)
```
